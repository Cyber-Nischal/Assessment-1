Task 6: Adding HDI Category Column

Creating a new column HDI Category that classifies each country into one of the four official
Human Development Index groups. The classification should be based on the HDI value for the
year 2022.

```python
def classify_hdi(hdi):
    if hdi < 0.550:
        return 'Low'
    elif 0.550 <= hdi <= 0.699:
        return 'Medium'
    elif 0.700 <= hdi <= 0.799:
        return 'High'
    else:  # hdi >= 0.800
        return 'Very High'

hdi_2022_df['HDI_Category'] = hdi_2022_df['hdi'].apply(classify_hdi)
```

Task 3: Outlier Detection (Warning Free)

```python
# Make sure south_asia_df is a copy
south_asia_df = south_asia_df.copy()

# Function to detect outliers
def detect_outliers(series):
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return (series < lower_bound) | (series > upper_bound)

# Detect HDI and GNI per Capita outliers using .loc to avoid warning
south_asia_df.loc[:, 'hdi_outlier'] = detect_outliers(south_asia_df['hdi'])
south_asia_df.loc[:, 'gni_outlier'] = detect_outliers(south_asia_df['gross_inc_percap'])

# Mark any country as outlier if either HDI or GNI per Capita is an outlier
south_asia_df.loc[:, 'outlier'] = south_asia_df['hdi_outlier'] | south_asia_df['gni_outlier']

# Scatter Plot: GNI per Capita vs HDI
plt.figure(figsize=(12,7))

# Normal countries
plt.scatter(
    south_asia_df.loc[~south_asia_df['outlier'], 'gross_inc_percap'],
    south_asia_df.loc[~south_asia_df['outlier'], 'hdi'],
    color='blue',
    alpha=0.7,
    s=60,
    label='Normal'
)

# Outlier countries
plt.scatter(
    south_asia_df.loc[south_asia_df['outlier'], 'gross_inc_percap'],
    south_asia_df.loc[south_asia_df['outlier'], 'hdi'],
    color='red',
    marker='D',
    s=80,
    label='Outlier'
)

# Annotate unique outlier countries only
for i, row in south_asia_df[south_asia_df['outlier']].drop_duplicates(subset=['country']).iterrows():
    plt.text(row['gross_inc_percap']*1.01, row['hdi']*1.01, row['country'], fontsize=10, weight='bold')

plt.xlabel('GNI per Capita')
plt.ylabel('HDI')
plt.title('GNI per Capita vs HDI with Outliers Highlighted (South Asia)', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend()
plt.tight_layout()
plt.show()

print("Outlier Countries in South Asia:")
print(south_asia_df[south_asia_df['outlier']][['country', 'hdi', 'gross_inc_percap']])

```

Task 5: Gap Analysis

```python
from matplotlib.patches import Patch

# Set a nicer style
sns.set_style("whitegrid")

# Ensure we work on a copy
south_asia_df = south_asia_df.copy()

# 1. Create GNI-HDI Gap
south_asia_df['gni_hdi_gap'] = south_asia_df['gross_inc_percap'] - south_asia_df['hdi']

# 2. Select latest year
latest_year = south_asia_df['year'].max()
gap_df = south_asia_df[south_asia_df['year'] == latest_year].copy()

# 3. Top 3 positive and negative gaps
top_positive = gap_df.sort_values('gni_hdi_gap', ascending=False).head(3)
top_negative = gap_df.sort_values('gni_hdi_gap').head(3)

# 4. Combine for plotting
plot_df = pd.concat([top_positive, top_negative])
plot_df['gap_type'] = plot_df['gni_hdi_gap'].apply(lambda x: 'Positive' if x > 0 else 'Negative')

# 5. Colors for each bar
positive_colors = sns.color_palette("Greens", 3)
negative_colors = sns.color_palette("Oranges", 3)
colors = positive_colors + negative_colors
color_dict = dict(zip(plot_df['country'], colors))

plt.figure(figsize=(10,6))

# 6. Plot bar chart using hue='country'
barplot = sns.barplot(
    data=plot_df,
    x='gni_hdi_gap',
    y='country',
    hue='country',
    palette=color_dict,
    dodge=False,
    errorbar=None,
    legend=False
)

# Adding values inside bars
for i, row in enumerate(plot_df.itertuples()):
    xpos = row.gni_hdi_gap - 0.1 if row.gni_hdi_gap > 0 else row.gni_hdi_gap + 0.02
    ha = 'right' if row.gni_hdi_gap > 0 else 'left'
    barplot.text(
        xpos,
        i,
        f"{row.gni_hdi_gap:.2f}",
        color='black',
        va='center',
        ha=ha,
        fontweight='bold'
    )

# 7. Labels, title, grid, spines
plt.xlabel('GNI-HDI Gap', fontsize=12)
plt.ylabel('Country', fontsize=12)
plt.title('Top 3 Positive and Negative GNI-HDI Gaps (South Asia)', fontsize=14, fontweight='bold')
plt.grid(axis='x', linestyle='--', alpha=0.7)
sns.despine(top=True, right=True)

# 8. Manual legend
legend_elements = [
    Patch(facecolor=positive_colors[1], label='Top Positive Gap'),
    Patch(facecolor=negative_colors[1], label='Top Negative Gap')
]
plt.legend(handles=legend_elements, loc='lower right')

plt.tight_layout()
plt.show()

# 9. Display tables
print("Top 3 Positive Gaps (GNI >> HDI):")
print(top_positive[['country', 'gross_inc_percap', 'hdi', 'gni_hdi_gap']])
print("\nTop 3 Negative Gaps (HDI >> GNI):")
print(top_negative[['country', 'gross_inc_percap', 'hdi', 'gni_hdi_gap']])
```

Task 5: HDI Disparity

```python
# Compute HDI range (max - min) for each region
sa_hdi_range = south_asia_df['hdi'].max() - south_asia_df['hdi'].min()
me_hdi_range = middle_east_df['hdi'].max() - middle_east_df['hdi'].min()

# Compute HDI mean and std for each region
sa_hdi_mean = south_asia_df['hdi'].mean()
sa_hdi_std = south_asia_df['hdi'].std()
me_hdi_mean = middle_east_df['hdi'].mean()
me_hdi_std = middle_east_df['hdi'].std()

# Compute Coefficient of Variation (CV)
sa_hdi_cv = sa_hdi_std / sa_hdi_mean
me_hdi_cv = me_hdi_std / me_hdi_mean

# Display results
print("HDI Disparity for South Asia:")
print(f"  Range (max - min): {sa_hdi_range:.3f}")
print(f"  Coefficient of Variation (CV): {sa_hdi_cv:.3f}")

print("\nHDI Disparity for Middle East:")
print(f"  Range (max - min): {me_hdi_range:.3f}")
print(f"  Coefficient of Variation (CV): {me_hdi_cv:.3f}")

# Identify which region exhibits more variation
if sa_hdi_cv > me_hdi_cv:
    print("\nSouth Asia exhibits more variation in HDI (higher CV).")
elif me_hdi_cv > sa_hdi_cv:
    print("\nMiddle East exhibits more variation in HDI (higher CV).")
else:
    print("\nBoth regions exhibit similar variation in HDI.")
```

Task 6: Correlation Analysis:

```python
# Metrics to correlate with HDI
metrics = ['gender_development', 'life_expectancy']
metric_labels = ['Gender Development Index', 'Life Expectancy Index']

# Function to compute correlations and plot scatter
def correlation_analysis(df, region_name):
    print(f"\nCorrelation Analysis for {region_name}:\n")
    for metric, label in zip(metrics, metric_labels):
        corr = df['hdi'].corr(df[metric])
        print(f"HDI vs {label}: correlation = {corr:.3f}")

        # Scatter plot with trendline
        plt.figure(figsize=(6,4))
        sns.regplot(x=metric, y='hdi', data=df, ci=None, scatter_kws={'s':50, 'color':'skyblue'}, line_kws={'color':'red'})
        plt.xlabel(label)
        plt.ylabel('HDI')
        plt.title(f"{region_name}: HDI vs {label} (corr={corr:.3f})", fontsize=12, fontweight='bold')
        plt.tight_layout()
        plt.show()

# Run for South Asia
correlation_analysis(south_asia_df, "South Asia")

# Run for Middle East
correlation_analysis(middle_east_df, "Middle East")
```

Task 7: Outlier Detection

```python
# Metrics to check for outliers
metrics = ['hdi', 'gross_inc_percap']
metric_labels = ['HDI', 'GNI per Capita']

def detect_outliers_clean(df, region_name):
    print(f"\nOutlier Detection for {region_name}:\n")

    for metric, label in zip(metrics, metric_labels):
        Q1 = df[metric].quantile(0.25)
        Q3 = df[metric].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        # Identify outliers
        outliers = df[(df[metric] < lower_bound) | (df[metric] > upper_bound)]
        print(f"{label}: {len(outliers)} outlier(s)")
        if not outliers.empty:
            print(outliers[['country', 'year', metric]])

        # Horizontal scatter plot for cleaner labels
        plt.figure(figsize=(8,6))
        # Normal points
        normal_points = df[~df.index.isin(outliers.index)]
        sns.scatterplot(x=normal_points[metric], y=normal_points['country'], color='skyblue', s=80, label='Normal')
        # Outliers
        if not outliers.empty:
            sns.scatterplot(x=outliers[metric], y=outliers['country'], color='red', s=100, label='Outlier')

        plt.xlabel(label)
        plt.ylabel('Country')
        plt.title(f"{region_name}: {label} Outlier Detection", fontsize=14, fontweight='bold')
        plt.legend()
        plt.tight_layout()
        plt.show()

# Detect outliers for South Asia
detect_outliers_clean(south_asia_df, "South Asia")

# Detect outliers for Middle East
detect_outliers_clean(middle_east_df, "Middle East")
```
